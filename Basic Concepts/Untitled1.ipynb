```python
# all rows where miles>10 and print only miles col
df2 = df.loc[df["MILES*"] > 10,["MILES*"]]
df2
```

```python
# first 5 trips starting from newyork(display only miles col)
df.loc[df["START*"]=="New York",["MILES*"]].head()
```

```python
# first 5 trips starting from cary or morrisville
df[df["START*"].isin(["Cary","Morrisville"])].head()
```

```python
# Find all trips with distance > 10 miles and originating from Cary and Morris.
# dispaly all col values
df.loc[(df["MILES*"] > 10) & (df["START*"].isin(["Cary","Morrisville"]))].head(n=3)
```

```python
# reset the above index
df3=df.loc[(df["MILES*"] > 10) & (df["START*"].isin(["Cary","Morrisville"]))].head(n=3)
df3.reset_index(inplace=True, drop=True)
df3
```

```python
# sort miles in descending order
df.sort_values(by=["MILES*"],ascending=False).head(5)
```

```python
# sort  start location in ascending and miles in descending order
df.sort_values(by=["START*", "MILES*"],ascending=[True,False]).head()
```

```python
# adding new col miles_cat
import numpy as np
df["miles_cat"] = np.where(df["MILES*"] > 5,"Long Trip","Short Trip")
df
```
